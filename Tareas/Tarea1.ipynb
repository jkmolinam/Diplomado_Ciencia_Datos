## **Aprendizaje supervisado**
El aprendizaje supervisado tiene como objetivo encontrar una funci√≥n o aplicaci√≥n medible
$$\hat{f}_{n}: X ‚Üí Y$$
Calculada a partir de muestras de entrenamiento de la forma 
$$(X_{i}, Y_{i})$$
para $i =1, \cdots , n$, de tal modo que la expresi√≥n $\hat{f}_{n}$ esta cercana a $Y$. Para ello, se asocia a los conjuntos $X, Y$ a las $\sigma$- √°lgebras $B(X), B(Y)$, tal que la medibilidad de las variables X, Y
est√° bien definida.


*   Regla de decisi√≥n: Es una funci√≥n medible $f:X\rightarrow Y$ 
*   Funci√≥n de p√©rdida: Es una funci√≥n medible $L: Y \times Y \rightarrow \mathbb{R}^{+} \cup 0$. La aplicaci√≥n de la regla de decisi√≥n $f$ a alg√∫n $X$ se define como $L(Y, f(X))$

 $L(x,y)=(x-y)^{2}$ que se denomina p√©rdida cuadr√°tica.

*   Si el problema es de clasificaci√≥n, utilizamos la funci√≥n de p√©rdida 0-1 en donde se toma el valor 0 si la decisi√≥n es correcta y 1 si la decisi√≥n es incorrecta.
